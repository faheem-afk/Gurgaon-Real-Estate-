### Initiating Data Cleaing

```python
df.drop(columns = ['link', 'property_id'], inplace=True) ## dropping the link and property_id columns
```

```python
df.columns
```

```python
df.head()
```

```python
df.rename(columns={'area':'price_per_sqft'}, inplace=True)
```

```python
df.society.unique()
```

```python
pattern = re.compile(r"\d+(\.\d+)?\s?★")
```

```python
df['society'] = df['society'].apply(lambda name: re.sub(pattern, '', str(name))).str.strip().str.lower() ## cleaning the society feature
```

```python
len(df['society'].unique())
```

```python
df['price'].value_counts()
```

```python
def lac_to_crore(price):
    if type(price) == float:
        return price
    else:        
        price_li = price.split()
        tag = price_li[-1].lower()
        if price == 'Price on Request':
            return np.nan
        if 'lac' in tag:
            crore_value  = float(price_li[0]) / 100
            return round(crore_value, 2)
        else:
            return round(float(price_li[0]), 2)
        
```

```python
df['price'] = df['price'].apply(lambda price: lac_to_crore(price))
```

```python
li = []
for i in df['price_per_sqft']:
    li.append(i)
len(li)
```

```python
pattern = re.compile(r"\b/sq.ft.")
def per_sqft(pattern, li):
    match = re.findall(pattern, str(li))
    return len(match)    

per_sqft(pattern, li) # The matching length of the array tells us that only sqft unit has been used.
```

```python
df['price_per_sqft'] = df['price_per_sqft'].apply(lambda price: str(price).split('/')[0].replace(',', '').replace('₹', '').strip()).astype('float')
```

```python
df['bedRoom'] = df['bedRoom'].apply(lambda room: room.split(' ')[0]).astype(int)
```

```python
df['bedRoom'].value_counts()
```

```python
df['bathroom'] = df['bathroom'].apply(lambda room: room.split(' ')[0]).astype(int)
```

```python
df['bathroom'].value_counts()
```

```python
df['balcony'] = df['balcony'].apply(lambda room: room.split(' ')[0] if room != 'No Balcony' else 0)
```

```python
df['balcony'].value_counts()
```

```python
df.fillna({'additionalRoom': 'not available'}, inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].apply(lambda x: x.lower())
```

```python
df['additionalRoom'].value_counts()
```

```python
np.where(pd.isna(df['floorNum']))[0]
```

```python
df['floorNum']
```

```python
df['floorNum'] = df['floorNum'].apply(
    lambda x: (
        0 if isinstance(x, str) and (x.split(' ')[0] == 'Ground' or x.split(' ')[0].lower() == 'lower')
        else -1 if isinstance(x, str) and x.split(' ')[0] == 'Basement'
        else re.sub(re.compile(r"\D+"), '', x.split(' ')[0]).strip() if isinstance(x, str)
        else x
    )
)
```
